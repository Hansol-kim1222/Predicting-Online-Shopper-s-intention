```python
df = pd.crosstab(data['VisitorType'], data['Revenue'])
df.div(df.sum(1).astype(float), axis=0).plot(kind = 'bar', stacked = True, figsize =(15, 5))
plt.title('Visitor Type & Revenue', fontsize = 30)
plt.show()
```
